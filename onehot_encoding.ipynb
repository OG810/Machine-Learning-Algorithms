```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('/cars.csv')
print(df) ##prints the dataset
print(df['brand'].value_counts()) ## prints all th categories and their frequienies under brand column
print(df['brand'].nunique())  ## prints the total unique categories under brand column
print(df['owner'].value_counts())
print(df['owner'].nunique())
print(df['fuel'].value_counts())
print(df['fuel'].nunique())
```

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(23) #This line sets the random seed for NumPy's random number generator. Setting the seed ensures that if you run the code again, you'll get the same random numbers. This is useful for reproducibility.
mu_vec1=np.array([0,0,0]) ## centralizing the mean.
cov_mat1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)
## Here, we define the mean vector (mu_vec1) and covariance matrix (cov_mat1) for the first set of samples. The np.random.multivariate_normal function generates random samples from a multivariate normal distribution. In this case, it generates 20 samples with mean mu_vec1 and covariance matrix cov_mat1
```

```python
df = pd.DataFrame(class1_sample,columns=['feature1','feature2','feature3'])
df['target'] = 1
## This code creates a Pandas DataFrame (df) using the samples from class1_sample. It has three columns ('feature1', 'feature2', 'feature3') representing different features, and a new column 'target' is added with all values set to 1. This 'target' column is often used in machine learning to represent the class label (in this case, class 1)
mu_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20)
## Similar to the first set of samples, this code generates another set of 20 samples (class2_sample) with a different mean vector (mu_vec2) and covariance matrix (cov_mat2). Another DataFrame (df1) is created with the same features but a 'target' column with all values set to 0.

df1 = pd.DataFrame(class2_sample,columns=['feature1','feature2','feature3'])

df1['target'] = 0

df = df.append(df1,ignore_index=True)

df = df.sample(40)
     ## This line appends the rows of df1 to df to create a single DataFrame with 40 samples. The ignore_index=True parameter reindexes the resulting DataFrame.

df.head()
```

```python

import plotly.express as px
#y_train_trf = y_train.astype(str)
fig = px.scatter_3d(df, x=df['feature1'], y=df['feature2'], z=df['feature3'],
              color=df['target'].astype('str'))
fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))

fig.show()
```

```python

# Step 1 - Apply standard scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df.iloc[:,0:3] = scaler.fit_transform(df.iloc[:,0:3])
```

```python
# Step 2 - Find Covariance Matrix
covariance_matrix = np.cov([df.iloc[:,0],df.iloc[:,1],df.iloc[:,2]])
print('Covariance Matrix:\n', covariance_matrix)
```

```python
# Step 3 - Finding EV and EVs
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python

eigen_values
```

```python
pc = eigen_vectors[0:2]
pc
```

```python
transformed_df = np.dot(df.iloc[:,0:3],pc.T)
# 40,3 - 3,2
new_df = pd.DataFrame(transformed_df,columns=['PC1','PC2'])
new_df['target'] = df['target'].values
new_df.head()
```

```python
new_df['target'] = new_df['target'].astype('str')
fig = px.scatter(x=new_df['PC1'],
                 y=new_df['PC2'],
                 color=new_df['target'],
                 color_discrete_sequence=px.colors.qualitative.G10
                )

fig.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig.show()
```

```python
## by default date and time column in dataset acts like a string. We have to convert to date and time datatype
## so that we can apply all those date and time features.

date['date']=pd.to_datetime(date['date'])
date.info()
```

```python
## extratcing the year from the date
date['date_year'] = date['date'].dt.year

date.sample(5)
```

```python
## extracting the month number
date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
##extracting the month name
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
##Extract Days
date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week ie whether it is 3rd day of week or 6th day of the week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```
